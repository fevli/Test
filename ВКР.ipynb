##Удаляем выбросы.

```python
for col in df.columns:
    q75,q25 = np.percentile(df.loc[:,col],[75,25])
    intr_qr = q75-q25

    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)

    df.loc[df[col] < min,col] = np.nan
    df.loc[df[col] > max,col] = np.nan
```

##Модели

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.pipeline import Pipeline
```

```python
y_mur = df_norm['Модуль упругости при растяжении, ГПа']
y_pr = df_norm['Прочность при растяжении, МПа']
x_mur = df_norm.drop(['Модуль упругости при растяжении, ГПа'], axis=1)
x_pr = df_norm.drop(['Прочность при растяжении, МПа'], axis=1)

X_train_mur, X_test_mur, y_train_mur, y_test_mur = train_test_split(x_mur, y_mur, test_size=0.3, random_state=1)
X_train_pr, X_test_pr, y_train_pr, y_test_pr = train_test_split(x_pr, y_pr, test_size=0.3, random_state=1)
```

Линейная регрессия

```python
lrg_mur = LinearRegression()
lrg_mur.fit(X_train_mur, y_train_mur)
```

```python
y_pred_mur = lrg_mur.predict(X_test_mur)
```

```python
print('Coefficients: \n', lrg_mur.coef_)
print('Mean squared error: %.2f' % mean_squared_error(y_test_mur, y_pred_mur))
print('Mean absolute error: %.2f' % mean_absolute_error(y_test_mur, y_pred_mur))
print('Coefficient of determination: %.2f' % r2_score (y_test_mur, y_pred_mur))
print('Intercept: \n', lrg_mur.intercept_)
```

```python
result1 = pd.DataFrame({'Model':'lrg_mur', 'MAE':mean_absolute_error(y_test_mur, y_pred_mur),
                              'r2_score':r2_score (y_test_mur, y_pred_mur)}, index=['Модуль упругости при растяжении, ГПа'])
```

```python
lrg_pr = LinearRegression()
lrg_pr.fit(X_train_pr, y_train_pr)
```

```python
y_pred_pr = lrg_pr.predict(X_test_pr)
```

```python
print('Coefficients: \n', lrg_pr.coef_)
print('Mean squared error: %.2f' % mean_squared_error(y_test_pr, y_pred_pr))
print('Mean absolute error: %.2f' % mean_absolute_error(y_test_pr, y_pred_pr))
print('Coefficient of determination: %.2f' % r2_score (y_test_pr, y_pred_pr))
print('Intercept: \n', lrg_pr.intercept_)
```

```python
result2 = pd.DataFrame({'Model':'lrg_pr', 'MAE':mean_absolute_error(y_test_pr, y_pred_pr),
                              'r2_score':r2_score (y_test_pr, y_pred_pr)}, index=['Прочность при растяжении, МПа'])
```

Рекомендательная нейронная сеть для соотношения матрица-наполнитель

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

print(tf.__version__)
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
```

```python
x_mn = df.drop(['Соотношение матрица-наполнитель'], axis=1)
y_mn = df[['Соотношение матрица-наполнитель']]
X_train_mn, X_test_mn, y_train_mn, y_test_mn = train_test_split(x_mn, y_mn, test_size=0.3, random_state=1)
```

```python
X_train_mn_norm = normalizer.adapt(np.array(X_train_mn))
```

```python
def build_and_compile_model(normalizer):
    model = keras.Sequential([normalizer,
                layers.Dense(128, activation='relu'),
                layers.Dense(64, activation='relu'),
                layers.Dense(32, activation='relu'),
                layers.Dense(1)])
    model.compile(loss='mean_absolute_error',
                  optimizer = 'adam')
    return model
```

```python
model_mn = build_and_compile_model(normalizer)
```

```python
model_mn.summary()
```

```python
%%time
history_mn = model_mn.fit(
    X_train_mn,
    y_train_mn,
    batch_size = 64,
    epochs=150,
    verbose=1,
    validation_split = 0.2)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(history_mn.history['loss'])
plt.plot(history_mn.history['val_loss'])
plt.title('График потерь модели', size=12)
plt.ylabel('Средняя абсолютная ошибка', size=12)
plt.xlabel('Эпоха', size=12)
plt.legend(['loss', 'val_loss'], loc='best')
plt.show()
```

```python
pred_mn = model_mn.predict(np.array((X_test_mn)))
original_mn = y_test_mn.values
predicted_mn = pred_mn
```

```python
plt.figure(figsize=(10,5))
plt.title('Тестовые и прогнозные значения', size=12)
plt.plot(original_mn, color='blue', label = 'Тестовые значения')
plt.plot(predicted_mn, color='red', label = 'Прогнозные значения')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Рассеяние тестовых и прогнозных значений', size=15)
plt.scatter(original_mn, predicted_mn)
plt.xlabel('Тестовые значения', size=12)
plt.ylabel('Прогнозные значения', size=12)
plt.show()
```

```python
print(f'Model MAE: {model_mn.evaluate(X_test_mn, y_test_mn)}')
```
